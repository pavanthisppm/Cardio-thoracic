```python
import pandas as pd
import numpy as np
import sklearn
import warnings
warnings.filterwarnings("ignore")
```

```python
data = pd.read_csv("ThoraricSurgery.csv")
data.head(2)
```

```python
#To create dummies (categories)
diagnose = pd.get_dummies(data["diagnose"],drop_first=True)
diagnose.head(2)
```

```python
pain = pd.get_dummies(data["pain"],drop_first=True)
pain.head(2)
```

```python
Haemoptysis = pd.get_dummies(data["Haemoptysis"],drop_first=True)
Haemoptysis.head(2)
```

```python
short_breath  = pd.get_dummies(data["short_breath"],drop_first=True)
short_breath .head(2)
```

```python
cough = pd.get_dummies(data["cough"],drop_first=True)
cough.head(2)
```

```python
weekness = pd.get_dummies(data["weekness"],drop_first=True)
weekness.head(2)
```

```python
tumour_size = pd.get_dummies(data["tumour_size"],drop_first=True)
tumour_size.head(2)
```

```python
diabetes = pd.get_dummies(data["diabetes"],drop_first=True)
diabetes.head(2)
```

```python
breathing_difficulty = pd.get_dummies(data["breathing_difficulty"],drop_first=True)
breathing_difficulty.head(2)
```

```python
other_disease = pd.get_dummies(data["other_disease"],drop_first=True)
other_disease.head(2)
```

```python
smoking = pd.get_dummies(data["smoking"],drop_first=True)
smoking.head(2)
```

```python
asthma = pd.get_dummies(data["asthma"],drop_first=True)
asthma.head(2)
```

```python
data = pd.concat([data,diagnose, pain,	Haemoptysis,	short_breath,	cough,	weekness,	tumour_size,	diabetes,	breathing_difficulty,	other_disease,	smoking,	asthma],axis = 1)
data.head(2)
```

```python
data.drop(['id', 'diagnose', 'pain',	'Haemoptysis',	'short_breath',	'cough',	'weekness',	'tumour_size',	'diabetes',	'breathing_difficulty',	'other_disease',	'smoking',	'asthma'],axis=1,
                  inplace=True)
data.head()
```

```python
#Training and Testing data
X =  data.drop("survive",axis=1)
y =  data["survive"]
from sklearn.model_selection import train_test_split
```

```python
#Splitting into training set and testing set
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
from sklearn.linear_model import LogisticRegression
logmodel = LogisticRegression()
logmodel.fit(X_train,y_train)
```

```python
predictions = logmodel.predict(X_test)
```

```python
#Metrics
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,predictions)*100)
```
